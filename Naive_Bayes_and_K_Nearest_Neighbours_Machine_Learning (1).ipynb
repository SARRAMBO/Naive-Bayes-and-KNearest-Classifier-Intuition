#Naive Bayes Algorithm-
 Bayes' TheoremAt its heart, the algorithm is just an application of Bayes' Theorem. This theorem allows us to update our beliefs based on new evidence.The general formula is:$$P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}$$In the context of machine learning, we swap $A$ and $B$ for things we care about: $y$ (the target class, like "Spam" or "Not Spam") and $X$ (the features or data points).$$P(y|X) = \frac{P(X|y) \cdot P(y)}{P(X)}$$

#Parameters for Hyperparameter Tuning in Gaussian NB-

1. Var Smoothing- (The Variance Stabilizer)By default, Scikit-Learn sets var_smoothing to $10^{-9}$ ($1\times 10^{-9}$). This is the primary hyperparameter tuned during cross-validation.The Problem It Solves: Gaussian NB assumes your features follow a normal distribution (bell curve). If a sample in your test dataset is far away from the training data mean, its calculated probability can drop to absolute zero. Because Naive Bayes multiplies the probabilities of all features together, a single zero will zero out the entire calculation, destroying the model's ability to make a prediction.How It Works: var_smoothing artificially expands the variance of all features by adding a fraction of the maximum variance found in the dataset to every feature's variance.The Impact: It acts as a regularization parameter. Increasing var_smoothing flattens and widens the Gaussian curves. This makes the model less sensitive to extreme outliers and forces it to be more conservative (less confident) in its predictions.


2. Priors: (The Class Weights)By default, Scikit-Learn sets priors=None. This means the algorithm automatically calculates the prior probabilities directly from your training data based on class frequency. For example, if your training data contains 70% Class A and 30% Class B, the priors are automatically set to $P(A) = 0.7$ and $P(B) = 0.3$.Why Tune It? You manually pass an array of probabilities (e.g., priors=[0.5, 0.5]) if you want to override the dataset's natural distribution.When To Use It: * Imbalanced Datasets: If your training data is heavily skewed (e.g., 99% legitimate emails, 1% fraud), the model will have a massive bias toward predicting the majority class. Manually adjusting the priors to be equal ($0.5 / 0.5$) forces the model to penalize misclassifications of the minority class more severely.Domain Knowledge: If you know ahead of time that the real-world deployment environment will have a different class distribution than your training set, you can set the expected priors manually.



### Why Naive Bayes Doesn't Need Standardization

**1. It Relies on Probability, Not Distance**
Standardization is crucial for algorithms that calculate the distance between data points (like K-Nearest Neighbors or Support Vector Machines) or algorithms that use gradient descent (like Neural Networks). Because Naive Bayes is a probabilistic classifier based on Bayes' Theorem, it doesn't measure spatial distances between features.

**2. The "Naive" Independence Assumption**
Naive Bayes assumes that every feature is completely independent of the others. When it calculates the probability of a data point belonging to a certain class, it evaluates each feature individually and then multiplies those probabilities together. Because the features are never directly weighed against each other in a combined formula, a feature with a scale of 1,000–10,000 will not "overpower" a feature with a scale of 0–1.

**3. The Math Balances Out (Gaussian Naive Bayes)**
If you are using Gaussian Naive Bayes for continuous data, the algorithm calculates the mean and variance for each feature within each class.

* If you input raw data, it uses the raw mean and raw variance to calculate the probability density.
* If you standardize the data, the mean becomes 0 and the variance becomes 1, and the probability density shifts accordingly.
Ultimately, the *relative* likelihoods between different classes remain exactly the same. Scaling changes the absolute numbers but doesn't change the final classification decision.

---

### When Standardization Can Actually Break Naive Bayes

In some variations of Naive Bayes, standardizing your data is not just unnecessary—it's actively harmful.

* **Multinomial Naive Bayes:** This variation is heavily used for text classification (like spam filtering) and relies on discrete counts (e.g., how many times a word appears in a document). Standardizing this data usually introduces negative values. Multinomial Naive Bayes **cannot handle negative values** because a negative frequency count doesn't make mathematical sense.
* **Sparse Data:** Text data is usually represented in sparse matrices (mostly zeros). Standardizing forces the data to be centered around zero, which converts all those meaningful zeros into non-zero floats. This destroys the sparsity, demanding massive amounts of memory and completely ruining computational efficiency.

### Summary

Save your preprocessing steps! You can safely feed your raw, unscaled features directly into a Naive Bayes model. Save standardization for algorithms like KNN, SVM, PCA, or Logistic Regression.

## K-Nearest Neighbours-
Instance based learning model used for classification and regression.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.datasets import make_classification

X, y = make_classification(
    n_samples=1000, # 1000 observations
    n_features=3, # 3 total features
    n_redundant=1,
    n_classes=2, # binary target/label
    random_state=999
)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X
```

```python
from sklearn.neighbors import KNeighborsClassifier

#p = 1 (manhattan distance)
#p = 2 (euclidean distance)

#algorithm- ball tree ,kdtree,brute
#kdtree,ball tree will create a binary tree (Time complexity reduced wrt search)
#auto-

#weights-{uniform and distance}

knn = KNeighborsClassifier(n_neighbors=6,p=2,algorithm='auto',weights='uniform')
knn.fit(X_train,y_train)
#features that are randomly distributed in feature dimensions-euclidean distance
```

## Hyperparameters for KNN-
1)Use Euclidean for continuous, low-dimensional data, and use Manhattan for high-dimensional data, sparse data, or when your dataset has many outliers.


 2) K-D Tree: Best for Low Dimensions ($D < 20$)How it works: It recursively slices your data space into a grid of hyper-rectangles using straight, axis-aligned cuts.When to use it: When you have fewer than 20 features/dimensions.Why it fails later: As you add more dimensions, the volume of those rectangles grows exponentially. The algorithm ends up having to search through almost every single "box" to find the true nearest neighbor, completely defeating the purpose of the tree. This is the "curse of dimensionality."2. Ball Tree: Best for High Dimensions ($D \ge 20$)How it works: It organizes data into a nested series of overlapping hyper-spheres (balls) rather than boxes.When to use it: When you have 20 or more features/dimensions.Why it works better here: Spheres don't have corners. In high-dimensional space, calculating the distance between spheres allows the algorithm to rule out large chunks of data much more efficiently than checking the corners of axis-aligned boxes.

 3) n_neighbours - The number of neighbouring pts estimated for predicting outcome of a new scoring point.

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix

accuracy_score(y_test,y_pred)
classification_report(y_test,y_pred)
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
param_grid = {
    'n_neighbors': [i for i in range(1,11)],
    'metric': ['euclidean', 'manhattan'],  # Testing distance metrics
    'algorithm': ['kd_tree', 'ball_tree']  # Testing spatial tree algorithms
}

grid_search = GridSearchCV(
    estimator=knn,
    param_grid=param_grid,
    cv=5,          # 5-fold cross-validation
    scoring='accuracy',
    n_jobs=-1      # Use all available CPU cores to speed it up
)

print("Running GridSearchCV... testing all 20 combinations...")
grid_search.fit(X_train, y_train)
```

```python
# Output the best results
print("\n--- Best Model Found ---")
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Cross-Validation Accuracy: {grid_search.best_score_:.4f}")

# 8. Final evaluation on the unseen test set
test_accuracy = grid_search.score(X_test, y_test)
print(f"Final Test Set Accuracy: {test_accuracy:.4f}")
```
